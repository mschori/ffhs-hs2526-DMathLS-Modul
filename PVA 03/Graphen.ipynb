# Graphen
## Grundbegriffe
- Ecken/Knoten/Verticen = Punkte in einem Graph
- Kanten/Edges = Verbindungen zwischen den Ecken
- Endlicher Graph = hat eine endliche Anzahl von Ecken und Kanten
- Unendlicher Graph = hat eine unendliche Anzahl von Ecken und Kanten
- Isomorphismus = Zwei Graphen sind isomorph, wenn sie die gleiche Struktur haben, d.h. es gibt eine Eins-zu-eins-Korrespondenz zwischen ihren Knoten und Kanten
- Schlichter Graph: Enthält die Menge E weder Mehrfachkanten noch Schlinen, dann heisst der Graph schlichter Graph
- Adjazenz/Benachbart = Zwei Ecken sind adjazent/benachbart, wenn sie durch eine Kante verbunden sind --> u und v sind adjazent, wenn {u,v} ∈ E
- Inzidenz = Eine Kante ist inzident zu einer Ecke, wenn sie diese Ecke verbindet --> e ist inzident zu v, wenn v ∈ e
      - Eine Kante ist inzident zu einer anderen Kante, wenn sie einen gemeinsamen Endpunkt haben --> e1 und e2 sind inzident, wenn e1 ∩ e2 ≠ ∅
- N(v) = Die Menge aller Nachbarn einer Ecke v wird N(v) geschrieben --> N(v) = {u ∈ V | {u,v} ∈ E}
- Grad einer Ecke (Eckengrad) = Der Grad einer Ecke ist die Anzahl der Nachbarn --> d(v) = |N(v)|
- Isolierter Knoten = Ein Knoten mit Grad 0 --> d(v) = 0
- |V| = Ordnung des Graphen = Anzahl der Ecken
- |E| = Größe des Graphen = Anzahl der Kanten
- Bipartiter Graph = Ein Graph, dessen Knoten in zwei disjunkte Mengen aufgeteilt werden können, so dass keine Kante zwei Knoten derselben Menge verbindet
- Weg = Eine Folge von Knoten, bei der jede zwei aufeinanderfolgenden Knoten durch eine Kante verbunden sind
- Kreise = Ein Weg, der an seinem Startknoten endet und keine Kante mehr als einmal verwendet
- Zusammenhang = Ein Graph ist zusammenhängend, wenn es einen Pfad zwischen jedem Paar von Knoten gibt
- Pfad = Ein Weg, bei dem kein Knoten zweimal besucht wird
- Zyklus = Ein Weg, der an seiner Startecke endet
- Kreis = Ein Zyklus, bei dem keine Ecke zweimal besucht wird
- Eulerweg = Ein Weg, der jede Kante genau einmal benutzt
- Eulertour = Ein Eulerweg, der an seinem Startknoten endet

- Schleife = Eine Kante, die einen Knoten mit sich selbst verbindet
- Multigraph = Ein Graph, der mehrere Kanten zwischen den gleichen Knoten erlaubt
- Gerichteter Graph (Digraph) = Ein Graph, bei dem die Kanten eine Richtung haben
- Ungerichteter Graph = Ein Graph, bei dem die Kanten keine Richtung haben
- Teilgraph = Ein Graph, der aus einem anderen Graphen durch Entfernen von Knoten und/oder Kanten entsteht
- Baum = Ein zusammenhängender, azyklischer Graph
- Bipartiter Graph = Ein Graph, dessen Knoten in zwei disjunkte Mengen aufgeteilt werden können, so dass keine Kante zwei Knoten derselben Menge verbindet
- Gewichteter Graph = Ein Graph, bei dem den Kanten Gewichte (Kosten, Längen, etc.) zugeordnet sind
- Planarer Graph = Ein Graph, der so gezeichnet werden kann, dass sich keine Kanten schneiden
- Komplementärer Graph = Ein Graph, der alle Kanten enthält, die im ursprünglichen Graphen nicht vorhanden sind
- Subgraph = Ein Graph, der aus einem anderen Graphen durch Entfernen von Knoten und/oder Kanten entsteht

## Einfacher Graph
- Ein (einfacher) Graph ist ein Paar G = (V, E) mit
    - V (Ecken) = ist eine Menge und nicht leer: V ≠ ∅
    - E (Kanten) = ist eine Menge von zweielementigen Teilmengen von V

## Vollständige Graphen
- Ein vollständiger Graph ist ein einfacher Graph, in dem jede Ecke mit jeder anderen Ecke durch genau eine Kante verbunden ist
- Hat zusätzlich V n Elemente, so nennt man G den vollständigen Graphen der Ordnung n und schreibt G = K<sub>n</sub> --> z.B. K<sub>3</sub>, K<sub>4</sub>, K<sub>5</sub>
- Die Anzahl der Kanten in einem vollständigen Graphen K<sub>n</sub> ist gegeben durch die Formel: |E| = n(n-1)/2 = n<sup>2</sup> - n / 2

## Bipartite Graphen
- Ein bipartiter Graph ist ein einfacher Graph, dessen Ecken in zwei disjunkte Mengen V<sub>1</sub> und V<sub>2</sub> aufgeteilt werden können, so dass jede Kante eine Ecke aus V<sub>1</sub> mit einer Ecke aus V<sub>2</sub> verbindet
- Dann gilt
    - (i) V<sub>1</sub> ∪ V<sub>2</sub> = V
    - (ii) V<sub>1</sub> ∩ V<sub>2</sub> = ∅
    - (iii) Für jede Kante von G gilt: V<sub>1</sub> geschnitten e ist nicht leer und V<sub>2</sub> geschnitten e ist nicht leer --> V<sub>1</sub> ∩ e ≠ ∅ und V<sub>2</sub> ∩ e ≠ ∅
- Wenn alle Ecken mit allen Ecken der anderen Menge verbunden sind, nennt man dies vollständig Bipartit --> K<sub>m,n</sub> (m = Anzahl der Ecken in V<sub>1</sub>, n = Anzahl der Ecken in V<sub>2</sub>)

## Weg, Kreis, Zusammenhang
- Ein Weg in einem Graphen G = (V, E) ist eine Folge von V mit der Eigenschaft, dass die Ecken durch Kanten verbunden sind
- Dabei kann eine Ecke mehrmals vorkommen, aber keine Kante mehr als einmal
- Sind alle Ecken eines Weges paarweise verschieden, also keine Ecke wird doppelt besucht, so nennt man ihn Pfad
- Hört der Weg dort auf, wo man angefangen hat, spricht man von einem Zyklus
- Wird bei einem Zyklus keine Ecke doppelt besucht, so nennt man dies Kreis

## Kantenanzahl und Eckengrade
- In jedem Graph (V, E) gilt: 2 * |E| = Summe über alle v ∈ V von d(v)
- D.h. die Summe aller Eckengrade ist immer gerade, da jede Kante zu genau zwei Eckgraden beiträgt
- Daraus folgt, dass die Anzahl der Ecken ungeraden Grades immer gerade ist

## Zusammenhang
- Wenn die Graphen nicht geteilt werden können, handelt es sich um einen zusammenhängenden Graphen
- Erreichbarkeit:
     - B ist von A aus erreichbar, wenn es einen Weg von A nach B gibt = A ~ B
     - B ist von A aus nicht erreichbar, wenn es keinen Weg von A nach B gibt = A !~ B
     - v ~ v gilt für alle, da jeder Punkt von sich selbst aus erreichbar ist (Reflexivität)
     - v ~ w <=> w ~ v (Symmetrie)
     - v1 ~ v2 und v2 ~ v3 => v1 ~ v3 (Transitivität)
     - Bedeutete auch bei Graphen gibt es eine Äquivalenzrelation => Reflexiv, Symmetrisch, Transitiv
- Zusammenhangskomponente: eine Zusammenhangskomponente ist eine maximale zusammenhängende Teilmenge von V
- Ein Graph ist zusammenhängend, wenn es nur eine Zusammenhangskomponente gibt
- Brücke: Eine Brücke ist eine Kante, deren Entfernung die Anzahl der Zusammhangskomponenten ehöhen würde

## Speichern von Graphen
- Adjazenzmatrix: Eine Adjazenzmatrix ist eine quadratische Matrix, die die Verbindungen zwischen den Ecken eines Graphen darstellt
    - Für einen Graphen G = (V, E) mit n Ecken ist die Adjazenzmatrix A eine n x n Matrix, wobei A[i][j] = 1, wenn es eine Kante zwischen Ecke i und Ecke j gibt, und A[i][j] = 0, wenn es keine Kante gibt
    - Für ungerichtete Graphen ist die Adjazenzmatrix symmetrisch
    - Für gerichtete Graphen ist die Adjazenzmatrix nicht unbedingt symmetrisch
    - Der Speicherbedarf für die Adjazenzmatrix beträgt O(n<sup>2</sup>), wobei n die Anzahl der Ecken ist
    - Vorteil: Schneller Zugriff auf Kanteninformationen = O(1)
    - Vorteil: Kanten sind schnell hinzugefügt oder entfernt = O(1)
    - Nachteil: Ineffizient für dünn besetzte Graphen (viele Nullen)
- Listen von Nachbarn: Eine Liste von Nachbarn ist eine Sammlung von Listen, wobei jede Liste die Nachbarn einer bestimmten Ecke enthält
    - Für einen Graphen G = (V, E) mit n Ecken ist die Liste der Nachbarn eine Liste von n Listen, wobei die i-te Liste die Nachbarn der i-ten Ecke enthält
    - Der Speicherbedarf für die Liste der Nachbarn beträgt O(n + m), wobei n die Anzahl der Ecken und m die Anzahl der Kanten ist
    - Vorteil: Effizienter Speicherbedarf für dünn besetzte Graphen
    - Nachteil: Langsamerer Zugriff auf Kanteninformationen = O(|E|) im schlimmsten Fall
    - Nachteil: Kanten sind langsamer hinzugefügt oder entfernt = O(|E|) im schlimmsten Fall

## Eulersche Graphen
- Ein Graph heisst eulerscher Graph, wenn es einen geschlossenen Weg gibt, der jede Kante genau einmal benutzt => nennt man auch Eulerweg
- Wenn bei einem Eulerweg Start- und Endpunkt identisch sind, nennt man dies Eulertour (Eulerkreis wäre nicht korrekt, da bei einem Kreis keine Ecke doppelt vorkommen darf)
- Voraussetzungen für einen Eulerweg:
    - Jede Ecke muss geraden Grad haben (Grad = Anzahl der Kanten, die an der Ecke anliegen)
    - Der Graph muss zusammehängend sein
    - Wenn die obigen beiden Voraussetzungen erfüllt sind, gibt es eine Eulertour
- Daher ist das Problem des Königsberger Brückenproblems unlösbar, da es Ecken mit ungeradem Grad gibt

## Zusammenhangskomponenten und Bäume

Eine Zusammenhangskomponente eines ungerichteten Graphen ist eine maximale Teilmenge von Knoten, in der jeder Knoten durch einen Pfad mit jedem anderen verbunden ist.

Algorithmisch: Man kann Zusammenhangskomponenten mit einer Tiefensuche (DFS) oder Breitensuche (BFS) bestimmen.
```python
def find_components(graph):
    visited = set()
    components = []

    def dfs(node, comp):
        visited.add(node)
        comp.append(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs(neighbor, comp)

    for node in graph:
        if node not in visited:
            comp = []
            dfs(node, comp)
            components.append(comp)

    return components
```

**Was ist ein Baum?**
Ein Baum ist ein zusammenhängender, zyklenfreier Graph. Ein Graph mit n Knoten ist genau dann ein Baum, wenn er n - 1 Kanten hat und zusammenhängend ist.

**Wichtige Eigenschaften von Bäumen**
- Jeder Baum mit n Knoten hat genau n - 1 Kanten.
- Es gibt zwischen zwei Knoten genau einen einfachen Pfad.
- Entfernt man eine Kante, entstehen zwei Zusammenhangskomponenten.

**Formel von Cayley**
Die Anzahl der verschiedenen Bäume, die man auf einer gegebenen Knotenmenge mit n Knoten bilden kann, ist:

𝑛<sup>𝑛−2</sup>

Beispiel:
Wie viele verschiedene Bäume gibt es auf 4 Knoten?

4<sup>4−2</sup> = 4<sup>2</sup> = 16

## Hamiltonsche Graphen

**Definition**
Ein Hamilton-Kreis ist ein geschlossener Weg, der jede Ecke genau einmal besucht. Ein Graph ist Hamiltonsch, wenn er einen solchen Kreis enthält.

**Unterschied zu Euler-Tour**

| Eigenschaft     | 	Euler-Tour       | Hamilton-Kreis             |
|-----------------|-------------------|----------------------------|
| Fokus           | 	Kanten           | 	Ecken                     |
| Jede Kante/Ecke | 	genau einmal     | 	genau einmal              |
| Voraussetzung   | 	Gradbedingungen  | 	keine einfache Bedingung  |

**Beispiel: Hamilton-Kreis**

Ein vollständiger Graph 𝐾<sub>4</sub> mit 4 Knoten ist Hamiltonsch. Ein möglicher Hamilton-Kreis: A → B → C → D → A

**Hinweis zur Aufgabe 3.5**

Wenn ein Knoten Grad 2 hat, müssen beide angrenzenden Kanten im Hamilton-Kreis enthalten sein. Das hilft beim Ausschluss von Kandidaten.

**Beziehung zum Travelling Salesman Problem (TSP)**

- TSP sucht den kürzesten Hamilton-Kreis in einem bewerteten Graphen.
- TSP ist NP-schwer, Hamilton-Kreis-Erkennung ist NP-vollständig.

## Planare Graphen

**Definition**
Ein Graph ist planar, wenn er ohne Kreuzungen in der Ebene gezeichnet werden kann.

**Eulersche Polyederformel**
Für zusammenhängende planare Graphen gilt:

𝑝 − 𝑞 + 𝑠 = 2

- p: Anzahl der Ecken
- q: Anzahl der Kanten
- s: Anzahl der Länder (Flächen)

Beispiel:

Ein Graph mit 6 Ecken und 9 Kanten:

6 − 9 + 𝑠 = 2 ⇒ 𝑠 = 5

**Kuratowski-Kriterium**

Ein Graph ist nicht planar, wenn er eine Unterteilung von 𝐾<sub>5</sub> oder 𝐾<sub>3,3</sub> enthält.

**Grenzwerte für Planarität**
- Für schlichte planare Graphen gilt: 𝑞 ≤ 3𝑝 − 6
- Für bipartite planare Graphen ohne Dreiecke: 𝑞 ≤ 2𝑝 − 4
